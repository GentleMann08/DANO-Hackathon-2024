```python
x_train = np.array(used_df['internet']).reshape(-1, 1)
y_train = np.array(used_df['lnwage']).reshape(-1, 1)

scaler = MinMaxScaler()
y_train_scaled = scaler.fit_transform(y_train)

def positive_transform(y):
    return np.log1p(y)

def positive_inverse_transform(y):
    return np.expm1(y)

model = TransformedTargetRegressor(
    regressor=LinearRegression(),
    func=positive_transform,
    inverse_func=positive_inverse_transform
)

model.fit(x_train, y_train)

model = TransformedTargetRegressor(
    regressor=LinearRegression(),
    func=positive_transform,
    inverse_func=positive_inverse_transform
)
model.fit(x_train, y_train_scaled)

```

```python
features = ['male', 'internet', 'educ']
target = 'lnwage'

X = used_df[features]
y = used_df[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)

r2 = r2_score(y_test, y_pred)

print(f'MSE (Среднеквадратическая ошибка): {mse}')

print(f'R² (Коэффициент детерминации): {r2}')

# new_property = [[0, 0, 0]]

# predicted_price = model.predict(new_property)

# print(f'Предсказанная заработная плата: {predicted_price[0]}')
```
